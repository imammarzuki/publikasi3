```python
import pandas as pd
from sklearn.preprocessing import normalize
from scipy.cluster.hierarchy import linkage, dendrogram
from scipy.spatial.distance import pdist
import matplotlib.pyplot as plt

# Load data
file_path = 'dataset3.xlsx'  # Ganti dengan path file Anda jika berbeda
df = pd.read_excel(file_path, sheet_name='Sheet1')

# Ambil hanya kolom data bulanan
monthly_data = df.iloc[:, 2:]

# Normalisasi dengan cosine
normalized_data = normalize(monthly_data, norm='l2')

# Hitung cosine distance dan linkage
cosine_distances = pdist(normalized_data, metric='cosine')
linked = linkage(cosine_distances, method='weighted')

# Plot dendrogram
plt.figure(figsize=(14, 18))
dendrogram(
    linked,
    labels=df['Province'].values,
    orientation='right',
    leaf_font_size=22  # Ukuran font provinsi diperbesar
)
# plt.title('Hierarchical Clustering Dendrogram (Rotated)', fontsize=20)
plt.xlabel('Cosine Distance', fontsize=22)
plt.ylabel('Province', fontsize=22)
plt.xticks(fontsize=22)
plt.tight_layout()
plt.show()
```
